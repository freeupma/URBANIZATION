While it has spurred economic growth and cultural diversity, urbanization also presents challenges related to inequality, environmental sustainability, and infrastructure development. Careful urban planning and policy measures are essential to harness the benefits of urbanization while addressing its associated challenges.

#1.5) Understanding the question
This dataset is named as the overall_poverty_est. It contains records under 7 attributes.
Below are the various variables in our dataset:

1. **residence_county** : the specific geographic region or administrative area within a country where an individual or a group of individuals live or have their primary place of residence
2. **Headcount Rate (%)** : a measure used in poverty analysis and social welfare assessment. It represents the proportion of a population that falls below the poverty line or threshold.  it indicates the percentage of people in a given area or group who are considered poor based on a specific poverty criterion.
3. **Distribution of the Poor (%)** :  refers to the way in which individuals or households living in poverty are spread across different income or wealth groups within a population. It provides insight into the concentration of poverty among various segments of the population, such as urban and rural areas or different socioeconomic classes.
4. **Poverty Gap (%)** : quantifies the depth of poverty experienced by individuals or households below the poverty line. It calculates the average income or resources needed to bring all the poor individuals or households up to the poverty threshold. A higher poverty gap percentage indicates a greater level of poverty severity.
5. **Severity of Poverty (%)** : assesses the intensity or severity of poverty among the poor population. It takes into account not only the headcount rate (the proportion of poor people) but also the depth of poverty experienced by them. A higher severity of poverty percentage indicates that the poor are further below the poverty line, implying a more acute level of poverty.
6. **Population (ths)** : It represent the total population of a region, country, or area in thousands. It provides the total count of individuals living in a specific geographic location.
7. **Number of Poor (ths)** :  represents the total count of individuals or households living in poverty within a specific area or group, expressed in thousands.
